## Visualization & Explanatory Analysis

### Pertanyaan 1: Analisis Kontribusi Penyewaan Akhir Pekan

```python
# Ekstrak tahun dan bulan dari kolom 'date'
df_day_cleaned['year'] = df_day_cleaned['date'].dt.year
df_day_cleaned['month'] = df_day_cleaned['date'].dt.month

# Menambahkan kolom untuk menunjukkan apakah hari tersebut merupakan akhir pekan
df_day_cleaned['day_of_week'] = df_day_cleaned['date'].dt.dayofweek
df_day_cleaned['is_weekend'] = df_day_cleaned['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# Menghitung total penyewaan di akhir pekan dan total keseluruhan
weekend_rentals = df_day_cleaned[df_day_cleaned['is_weekend'] == 1]['total_rentals'].sum()
total_rentals = df_day_cleaned['total_rentals'].sum()

# Menghitung kontribusi penyewaan akhir pekan dalam persentase
contribution = (weekend_rentals / total_rentals) * 100

# Menampilkan hasil
print(f"Kontribusi penyewaan akhir pekan: {contribution:.2f}%")

# Mengelompokkan data berdasarkan bulan dan status akhir pekan
monthly_rentals = df_day_cleaned.groupby(['month', 'is_weekend'])['total_rentals'].sum().reset_index()

# Visualisasi
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_rentals, x='month', y='total_rentals', hue='is_weekend', palette=['#007acc', '#ffa500'], linewidth=2.5)
plt.title('Kontribusi Penyewaan Akhir Pekan terhadap Total Penyewaan Bulanan', fontsize=16, fontweight='bold')
plt.xlabel('Bulan', fontsize=14)
plt.ylabel('Total Penyewaan', fontsize=14)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Jenis Hari', labels=['Hari Kerja', 'Akhir Pekan'], fontsize=12, title_fontsize=14, loc='upper left')
sns.despine()
plt.show()
```

**Insight : **
- Penyewaan sepeda di akhir pekan berkontribusi sekitar 28% dari total penyewaan sepeda. Ini menunjukkan bahwa akhir pekan memiliki peran signifikan dalam penggunaan layanan penyewaan sepeda.

## Analisis Lanjutan : RFM Analysis

Analysis Lanjutan dengan RFM Analysis untuk menilai hari-hari berdasarkan:

- Recency: Hari terakhir dengan penyewaan besar.
- Frequency: Frekuensi penyewaan sepeda per bulan.
- Monetary: Total penyewaan sepeda pada hari tersebut.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Memuat dataset 'day.csv'
day_df = pd.read_csv('/content/day.csv')

# Mengonversi kolom 'dteday' menjadi datetime
day_df['dteday'] = pd.to_datetime(day_df['dteday'])

# Recency: Menghitung jumlah hari sejak hari penyewaan terakhir
latest_date = day_df['dteday'].max()
day_df['days_since_last_rental'] = (latest_date - day_df['dteday']).dt.days

# Frequency: Menghitung frekuensi penyewaan sepeda per bulan
day_df['month_year'] = day_df['dteday'].dt.to_period('M')
frequency_df = day_df.groupby('month_year')['cnt'].count().reset_index()

# Monetary: Total penyewaan sepeda per hari
monetary_df = day_df.groupby('dteday')['cnt'].sum().reset_index()

# Menggabungkan hasil RFM Analysis
rfm_df = day_df[['dteday', 'days_since_last_rental', 'cnt']].copy()
rfm_df = rfm_df.groupby('dteday').agg({
    'days_since_last_rental': 'min',  # Recency: Hari terakhir penyewaan
    'cnt': ['sum', 'count']           # Monetary: Total penyewaan, Frequency: Frekuensi penyewaan
}).reset_index()

# Menamakan ulang kolom hasil aggregasi
rfm_df.columns = ['dteday', 'recency', 'monetary', 'frequency']

# Melihat hasil RFM Analysis
print(rfm_df.head())

# Visualisasi RFM Score
plt.figure(figsize=(10, 6))
plt.scatter(rfm_df['recency'], rfm_df['monetary'], c=rfm_df['frequency'], cmap='viridis', s=100, alpha=0.7)
plt.colorbar(label='Frequency')
plt.title('RFM Analysis: Recency vs Monetary vs Frequency')
plt.xlabel('Recency (Hari Terakhir Penyewaan)')
plt.ylabel('Monetary (Total Penyewaan)')
plt.grid(True)
plt.show()
```

**Penjelasan Analisis Lanjutan:**
- Recency: Diukur sebagai jumlah hari sejak hari penyewaan terakhir. Semakin kecil angka recency, semakin baru transaksi tersebut.
- Frequency: Menghitung jumlah transaksi penyewaan (kolom cnt) per hari.
- Monetary: Total jumlah penyewaan sepeda per hari (kolom cnt), di mana semakin besar nilai cnt, semakin tinggi volume transaksi pada hari itu.

**Hasil dan Visualisasi:**
- RFM Score dihasilkan dari tiga parameter (Recency, Frequency, dan Monetary).
- Visualisasi menggunakan scatter plot untuk melihat hubungan antara recency (hari sejak penyewaan terakhir), monetary (total penyewaan), dan frequency (frekuensi penyewaan). Titik pada plot diwarnai berdasarkan frekuensi, sehingga kita bisa mengidentifikasi hari-hari dengan aktivitas penyewaan yang lebih tinggi.

**Kesimpulan Analisis Lanjutan:**

Dari hasil analisis ini, kita bisa mengidentifikasi hari-hari dengan transaksi penyewaan tinggi, baik dari sisi total penyewaan maupun frekuensi penyewaan.
Hari-hari dengan Recency rendah dan Monetary tinggi menunjukkan hari-hari penyewaan terbaru yang aktif, sedangkan hari-hari dengan Recency tinggi memerlukan perhatian untuk memahami penurunan aktivitas.
